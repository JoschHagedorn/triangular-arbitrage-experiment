```python
import asyncio
from binance import AsyncClient, BinanceSocketManager
import traceback

from config_testnet import api_key, api_secret

from binance.exceptions import BinanceAPIException
import pandas as pd
import json
from collections import Counter
import time
```

```python
async def main():

    client = await AsyncClient.create(api_key=api_key, api_secret=api_secret, testnet=True)

    try:
        exchange_info = await client.get_exchange_info()
    except BinanceAPIException as e:
        print(e)
    symbol_info = exchange_info["symbols"]
    assets = []
    for d in symbol_info: 
        if d["status"] == "TRADING" and d["isSpotTradingAllowed"]:
            assets.append(d['baseAsset'])
            assets.append(d['quoteAsset'])

    await client.close_connection()

    # Count occurrences of each name
    asset_counts = Counter(assets)
    # Print names and their counts in descending order
    sorted_counts = sorted(asset_counts.items(), key=lambda x: x[1], reverse=True)

    for asset, count in sorted_counts:
        print(f'{asset}: {count}')
await main()
```
